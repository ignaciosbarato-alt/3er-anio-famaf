```python
def simular_juego_montecarlo(nsim):
    juegos_ganados = 0

    for i in range(nsim):
        U = random.random()


        if U < 1/2:
            w1 = random.random()
            w2 = random.random()
            x = w1 + w2

        else:
            w1 = random.random()
            w2 = random.random()
            w3 = random.random()

            x = w1 + w2 + w3

        if x >= 1:
            juegos_ganados += 1

        
    prob_estimada = juegos_ganados/nsim

    return prob_estimada
```

```python
simular_juego_montecarlo(100000000)
#vemos que a medida que aumenta el nsim mas se acerca la prob estimada a 2/3
```

```python
def simular_otro_juego(nsim):
    juegos_gan = 0

    for i in range(nsim):
        u = random.random()
        if u < 1/3:
            w1 = random.random()
            w2 = random.random()
            x = w1 + w2
        else: 

            w1 = random.random()
            w2 = random.random()
            w3 = random.random()
            x = w1 + w2 + w3


        if x <= 2:
            juegos_gan += 1

    return juegos_gan / nsim

```

```python
#estimacion de la g integrada

def montecarlo(g,nsim):
    integral = 0
    for i in range(nsim):
        integral += g(random.random())

    return integral/nsim
```

problema 7

```python
def suma_n():
    valor = 0
    i = 0
    while valor < 1:
        valor += random.random()
        i += 1

    return i

def media_n(nsim):
    media = 0
    for i in range(nsim):
        media += suma_n()
    
    return media/nsim
```

```python
media_n(110000)
```

```python
def prod_n():

    valor = 1
    i = 0

    while valor > math.exp(-3):
        i += 1
        valor *= random.random()

       

    return i - 1

def media_prod_n(nsim):
    media = 0
    for _ in range(nsim):
        media += prod_n()

    return media / nsim

```

```python
media_prod_n(1000000)
```
